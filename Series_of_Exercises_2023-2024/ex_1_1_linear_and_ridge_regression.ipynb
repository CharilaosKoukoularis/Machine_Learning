```python
import pandas as pd

wine_data = pd.read_csv('../../Matlab_Projects/Machine_Learning/ML2023-24-hwk1.csv', sep=';')
display(wine_data)
```

```python
from sklearn.preprocessing import StandardScaler
standardscaler = StandardScaler().fit(wine_data)#.drop(labels='quality',axis=1),wine_data['quality'])
normalized_data = pd.DataFrame(standardscaler.transform(wine_data))#.drop(labels='quality',axis=1)))
normalized_data.columns = wine_data.columns#.drop(labels='quality',axis=1).columns
display(normalized_data)
```

```python
print(normalized_data.mean(axis='index'),normalized_data.std(axis='index'))
print(wine_data.mean(axis='index'),wine_data.std(axis='index'))
```

```python
import matplotlib.pyplot as plt
plt.figure()
wine_data.hist(bins=30,figsize=(20,10))
plt.show()

plt.figure()
normalized_data.hist(bins=30,figsize=(20,10))
plt.show()
```

```python
import seaborn as sns

sns.pairplot(normalized_data, kind='reg', plot_kws={'line_kws':{'color':'red'}})
```

```python
X_train = normalized_data[:100].drop(columns='quality', axis=1)
y_train = normalized_data[:100]['quality']

X_test = normalized_data[100:150].drop(columns='quality', axis=1)
y_test = normalized_data[100:150]['quality']
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

l = [0, 10, 100, 200]

models = [Ridge(alpha=l[0]), Ridge(alpha=l[1]), Ridge(alpha=l[2]), Ridge(alpha=l[3])]

models[0].fit(X=X_train, y=y_train)
models[1].fit(X=X_train, y=y_train)
models[2].fit(X=X_train, y=y_train)
models[3].fit(X=X_train, y=y_train)

y_pred = [models[0].predict(X=X_train),
          models[1].predict(X=X_train),
          models[2].predict(X=X_train),
          models[3].predict(X=X_train),
          models[0].predict(X=X_test),
          models[1].predict(X=X_test),
          models[2].predict(X=X_test),
          models[3].predict(X=X_test),
]
rmse = [
    mean_squared_error(y_true=y_train, y_pred=y_pred[0]),
    mean_squared_error(y_true=y_train, y_pred=y_pred[1]),
    mean_squared_error(y_true=y_train, y_pred=y_pred[2]),
    mean_squared_error(y_true=y_train, y_pred=y_pred[3]),
    mean_squared_error(y_true=y_test, y_pred=y_pred[4]),
    mean_squared_error(y_true=y_test, y_pred=y_pred[5]),
    mean_squared_error(y_true=y_test, y_pred=y_pred[6]),
    mean_squared_error(y_true=y_test, y_pred=y_pred[7]),
]

rmse = pd.DataFrame([rmse[0:4], rmse[4:8]])
rmse.columns = ['Linear Regression Î» = 0',
                'Ridge Regression Î» = 10', 
                'Ridge Regression Î» = 100', 
                'Ridge Regression Î» = 200']

display(rmse)
```
